```python
#Multiple neurons
# Hidden layer
# Batch processing (4 inputs together)
# Forward pass only
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [1], [1], [0]])
```

```python
#hidden layer
W_hidden = np.array([
    [0.5, 0.4],   # Neuron 1 weights
    [0.3, 0.7]    # Neuron 2 weights
])
```

```python
b_hidden = np.array([0.1, 0.2])
```

```python
#Model ek hi time me sabka calculation karta hai using matrix operations.
# instead of:
# y=w⋅x+b(onesample)
# We do:
# Y=X⋅W+b
# Where:
# X = multiple inputs (batch)
# W = weights
# Y = multiple outputs together
Z = np.dot(X,W_hidden) + b_hidden
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
output = sigmoid(Z)
```

```python
print("Weighted Sum:", Z)
print("Final Output:", output)
```

```python
#output layer
W_out = np.array([[0.6],
                  [0.8]])

b_out = 0.05

Z2 = np.dot(output, W_out) + b_out
final = sigmoid(Z2)
```
