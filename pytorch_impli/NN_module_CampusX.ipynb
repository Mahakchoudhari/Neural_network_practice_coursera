```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):

    def __init__(self,num_feature):

        super().__init__()
        self.linear=nn.Linear(num_feature,1)
        self.sigmoid=nn.Sigmoid()

    def forward(self, fetures):
        out=self.linear(features)
        out=self.sigmoid(out)

        return out
```

```python
features = torch.rand(10,5)
```

```python
model=Model(features.shape[1])
```

```python
model(features)#it shows y_pred
```

```python
model.linear.weight
```

```python
model.linear.bias
```

```python
from torchinfo import summary

summary(model, input_size=(10, 5))
```

```python
class hidden_nn(nn.Module):
     def __init__(self,num_feature):

        super().__init__()
        self.linear1=nn.Linear(num_feature,3)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(3,1)
        self.sigmoid=nn.Sigmoid()

     def forward(self, fetures):
        out=self.linear1(features)
        out=self.relu(out)
        out=self.linear2(out)
        out=self.sigmoid(out)

        return out
```

```python
features = torch.rand(10,5)
```

```python
model2=hidden_nn(features.shape[1])
```

```python
model2(features)
```

```python
model2.linear1.weight
```

```python
from torchinfo import summary

summary(model2, input_size=(10, 5))
```
