```python
import numpy as np

# -------------------------------------------------
# DATASET
# -------------------------------------------------

# 2 features, 4 training examples
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# labels
Y = np.array([[0], [0], [0], [1]])

# -------------------------------------------------
# PARAMETERS (Neuron weights + bias)
# -------------------------------------------------

w = np.random.randn(2, 1)
b = 0

learning_rate = 0.1

# -------------------------------------------------
# ACTIVATION FUNCTION
# -------------------------------------------------

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# -------------------------------------------------
# TRAINING
# -------------------------------------------------

for i in range(1000):

    # Forward Propagation
    Z = np.dot(X, w) + b
    A = sigmoid(Z)

    # Loss
    m = X.shape[0]

    loss = -(1/m) * np.sum(
        Y * np.log(A) + (1 - Y) * np.log(1 - A)
    )

    # Backpropagation
    dZ = A - Y

    dw = (1/m) * np.dot(X.T, dZ)
    db = (1/m) * np.sum(dZ)

    # Gradient Descent
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Print loss
    if i % 100 == 0:
        print(f"Iteration {i}, Loss = {loss}")

# -------------------------------------------------
# FINAL PREDICTIONS
# -------------------------------------------------

print("\nPredictions:")

A = sigmoid(np.dot(X, w) + b)

predictions = (A > 0.5).astype(int)

print(predictions)
```
