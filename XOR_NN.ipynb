```python
import numpy as np
```

```python
X = np.array([
    [0, 0, 1, 1],
    [0, 1, 0, 1]
])
```

```python
Y = np.array([[0, 1, 1, 0]])
```

```python
X.shape
```

```python
Y.shape
```

```python
#hiiden layer
#2 hidden neurons
#2 input features
W1 = np.random.randn(2,2)
b1 = np.zeros((2,1))

#output layer
W2 = np.random.randn(1,2)
b2 = np.zeros((1,1))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def tanh_derivative(a):
    return 1 - np.square(a)
```

```python
learning_rate = 0.1
epochs = 10000
```

```python
m = X.shape[1]
```

```python
for i in range(epochs):
    #forward prop hidden layer
    Z1=np.dot(W1,X)+b1
    A1=np.tanh(Z1)

    #output layer
    Z2=np.dot(W2,A1)+b2
    A2=sigmoid(Z2)

    #loss 
    #numeric stability kai liye 1e-8 (small no.)taki log(0)ko avoid ker sake
    loss = - (1/m) * np.sum(
        Y * np.log(A2 + 1e-8) +
        (1 - Y) * np.log(1 - A2 + 1e-8)
    )

    #backward prop
    dZ2 = A2 - Y
    dW2 = (1/m) * np.dot(dZ2, A1.T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)

    dZ1 = np.dot(W2.T, dZ2) * tanh_derivative(A1)
    dW1 = (1/m) * np.dot(dZ1, X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)

    # UPDATE PARAMETERS
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1

    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2

    # PRINT LOSS
    if i % 1000 == 0:
        print(f"Epoch {i}, Loss = {loss:.4f}")
```

```python
# FINAL PREDICTIONS
print("\nFinal Predictions:\n")
print(A2)
```

```python
predictions = (A2 > 0.5).astype(int)
print("\nBinary Predictions:\n")
print(predictions)
```
