```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
X=torch.tensor([[1,1],
                [1,0],
                [0,1],
                [0,0]] ,dtype=torch.float32)
Y=torch.tensor([[0],
                [1],
                [1],
                [0]] ,dtype=torch.float32)
print(X.shape)
print(Y.shape)
```

```python
class NeuralNetwork(nn.Module):

    def __init__(self):
         super().__init__()

         self.fc1=nn.Linear(2,4)
         self.fc2=nn.Linear(4,1)
        
    def forward(self, x):
         x=self.fc1(x)
         x=torch.relu(x)
         x=self.fc2(x)
         x=torch.sigmoid(x)
        
         return x
```

```python
model=NeuralNetwork()
print(model)
```

```python
model=NeuralNetwork()
output=model(X)
print(output)
print(output.shape)
```

```python
loss=nn.BCELoss()
```

```python
cost=loss(output,Y)
print(cost)
```

```python
optimizer=optim.Adam(
    model.parameters(),
    lr=0.1
)
```

```python
for name , param in model.named_parameters():
    print(name)
    print(param)
    print("shape:", param.shape)
    print("------")
```

```python
optimizer.zero_grad()
epochs=2000
for epoch in range(epochs):
    #forward pass
    logits=model(X)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    if epoch % 500 == 0:
        print(f"Epoch {epoch} | Loss = {cost.item():.6f}")
```

```python
with torch.no_grad():
    output=model(X)
    
    prediction=(output>0.5).int()

    print("Probabilities:")
    print(output)

    print("\nPredictions:")
    print(prediction)

    print("\nActual:")
    print(Y)

    correct=(prediction == Y).int().sum()
    print(correct)
    accuracy=correct/Y.shape[0]
    print("accuracy is : " , accuracy.item()*100 ,"%")
```
