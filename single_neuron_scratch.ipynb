```python
import numpy as np
```

```python
#input define
x = np.array([2, 3])
#x1 = 2
#x2 = 3
```

```python
#w define
w = np.array([0.5, 0.8])
```

```python
#bias define
b = 1
```

```python
#neuron formula
z = np.dot(x, w) + b
```

```python
#activation formula
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Final Output
output = sigmoid(z)
```

```python
print("Weighted Sum:", z)
print("Final Output:", output)#neuron ki confidence ≈ 98%
```
